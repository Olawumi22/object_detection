```python
import cv2
import os
```

```python
cam = cv2.VideoCapture(r"C:\Users\HP\Downloads\The Airgun Show – epic night vision rat shoot, PLUS the Weihrauch HW98 Grey Laminate on test….mp4") #capture object to initialize the video
```

```python
try:
    if not os.path.exists('data'):
        os.makedirs('data')
except OSError:
    print('Error:creating directory of data')
currentframe= 0

while(cam.isOpened()):
    ret, frame=cam.read()       #ret means there is frame true, if not false, this is to ensure no empty frame
                                 #frame means frame

    if ret:
        name= './data/frame' + str(currentframe) 
        print('creating...'+ name)
        #cv2.imshow(name, frame)
        #cv2.waitKey(2)
        cv2.imwrite(name+'.jpg', frame)
        currentframe+= 1
    else:
        break
cam.release()
cv2.destroyAllWindows()
```
